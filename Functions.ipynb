1. What is the difference between a function and a method in Python?
- A function is a block of code that can be used anywhere. A method is a function that belongs to an object, like a string or list.

2. Explain the concept of function arguments and parameters in Python.
- Parameters are placeholders we write when defining a function. Arguments are the actual values we give when calling the function.

3. What are the different ways to define and call a function in Python?
- We can define a function using def or using a small one-line lambda function. We call a function by writing its name followed by parentheses.

4. What is the purpose of the return statement in a Python function?
- The return statement sends a result back from the function. Without it, the function only does something but doesn’t give a value to use later.

5. What are iterators in Python and how do they differ from iterables?
- An iterable is something we can loop over, like a list or string. An iterator is an object that gives items one by one when we ask for the next item.

6. Explain the concept of generators in Python and how they are defined.
- A generator is a special kind of function that gives values one at a time instead of all at once. It produces values only when needed.

7. What are the advantages of using generators over regular functions?
- Generators save memory because they don’t store all values at once. They are also faster when working with large amounts of data.

8. What is a lambda function in Python and when is it typically used?
- A lambda function is a small, one-line function without a name. It is used when we need a quick, simple function.

9. Explain the purpose and usage of the map() function in Python.
- map() is used to apply a function to every item in a list or other iterable, transforming all items according to the function.

10. What is the difference between map(), reduce(), and filter() functions in Python?

- map() changes each item using a function.

- filter() keeps only the items that meet a condition.

- reduce() combines all items into a single result.

```python
#1) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def SR(SF):
    EvenN=0
    for Numb in SF:
        if Numb%2==0:
            EvenN+=Numb
    return EvenN

SA=[2,4,5,6]
print(SR(SA))
```

```python
#4) Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

for n in range(1, 201):
    if is_prime(n):
        print(n, "is Prime")
    else:
        print(n, "is Not Prime")
```

```python
#5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b


fib = FibonacciIterator(12)
for num in fib:
    print(num)
```

```python
#6) 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(Num):
    for exp in range(Num + 1):
        yield 2 ** exp

for result in powers_of_two(5):
    print(result)
```

```python
#7) . Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()
```

```python
#10) Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, text))

string = "Hello World"
result = remove_vowels(string)
print(result)
```
